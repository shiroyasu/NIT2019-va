## 面積分 ##

空間内の曲線に関する線積分の類似として, スカラー場やベクトル場を「曲面に関して」積分することができる.
これを面積分という.

```
JupyterImageResolution = 72;
```

### 曲面の面積 ###

2変数ベクトル関数$\boldsymbol{r}(u,v)$で定義される曲面を$S$とする.

$S$上の点 $\mathrm{P}\boldsymbol{r}(u,v)$ から, パラメータを $\Delta u$, $\Delta v$ だけ動かした点 $\mathrm{P}_1\boldsymbol{r}(u+\Delta u,v)$, $\mathrm{P}_2\boldsymbol{r}(u+\Delta u,v+\Delta v)$, $\mathrm{P}_3\boldsymbol{r}(u,v+\Delta v)$ の4点を結ぶ座標曲線で囲まれる領域の面積を $\Delta S$ とおく.

```
funct71[x_,y_]:={x,y,x*y^2};
Dxfunct71[x_,y_]:=D[funct71[t,s],t]/.t->x/.s->y;
Dyfunct71[x_,y_]:=D[funct71[t,s],s]/.t->x/.s->y;
```

```
Show[
    ParametricPlot3D[funct71[x,y],{x,-1/2,0},{y,-1/2,0},Mesh->False],
    ParametricPlot3D[funct71[-0.4,t],{t,-0.4,-0.4+0.2}],ParametricPlot3D[funct71[-0.4+0.2,t],{t,-0.4,-0.4+0.2}],
    ParametricPlot3D[funct71[t,-0.4],{t,-0.4,-0.4+0.2}],ParametricPlot3D[funct71[t,-0.4+0.2],{t,-0.4,-0.4+0.2}],
    Graphics3D[{PointSize[0.02],Point[funct71[-0.4,-0.4]],Point[funct71[-0.4+0.2,-0.4]],Point[funct71[-0.4,-0.4+0.2]],Point[funct71[-0.4+0.2,-0.4+0.2]]}]
]
```

$\Delta u$, $\Delta v$ がともに微小量であれば, 上の領域の面積は $\overrightarrow{\mathrm{P}\mathrm{P}_1}$と$\overrightarrow{\mathrm{P}\mathrm{P}_3}$ を2辺とする平行四辺形の面積で近似できると考える\. 

```
Show[
    ParametricPlot3D[funct71[x,y],{x,-1/2,0},{y,-1/2,0},Mesh->False],
    ParametricPlot3D[funct71[-0.4,t],{t,-0.4,-0.4+0.2}],ParametricPlot3D[funct71[-0.4+0.2,t],{t,-0.4,-0.4+0.2}],
    ParametricPlot3D[funct71[t,-0.4],{t,-0.4,-0.4+0.2}],ParametricPlot3D[funct71[t,-0.4+0.2],{t,-0.4,-0.4+0.2}],
    Graphics3D[{PointSize[0.02],Point[funct71[-0.4,-0.4]],Point[funct71[-0.4+0.2,-0.4]],Point[funct71[-0.4,-0.4+0.2]],Point[funct71[-0.4+0.2,-0.4+0.2]],Arrow[{funct71[-0.4,-0.4],funct71[-0.4,-0.4]+0.2*Dxfunct71[-0.4,-0.4]}],Arrow[{funct71[-0.4,-0.4],funct71[-0.4,-0.4]+0.2*Dyfunct71[-0.4,-0.4]}]}]
]
```

ここで,
$$\overrightarrow{\mathrm{P}\mathrm{P}_1}
=\boldsymbol{r}(u+\Delta u,v)-\boldsymbol{r}(u,v)
=\frac{\boldsymbol{r}(u+\Delta u,v)-\boldsymbol{r}(u,v)}{\Delta u}\,\Delta u
\approx \frac{\partial \boldsymbol{r}}{\partial u}\,\Delta u
$$
$$
\overrightarrow{\mathrm{P}\mathrm{P}_3}
=\boldsymbol{r}(u,v+\Delta v)-\boldsymbol{r}(u,v)
=\frac{\boldsymbol{r}(u,v+\Delta v)-\boldsymbol{r}(u,v)}{\Delta v}\,\Delta v
\approx \frac{\partial \boldsymbol{r}}{\partial v}\,\Delta v
$$
なので, ベクトルの外積の性質から, 平行四辺形${\rm P}{\rm P}_1{\rm P}_2{\rm P}_3$の面積は
$$
\left|\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\right|\,\Delta u\Delta v
$$
で近似できることがわかる.

#### 面素 ####
曲面 $\boldsymbol{r}(u,v)$ に対し, 
$$
\left|\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\right|\,dudv
$$
を曲面の<b>面積素</b>, または<b>面素</b>といい, $dS$ と書く.

#### 事実 ####
パラメータ$(u,v)$が領域$D$を動くとき, 曲面 $\boldsymbol{r}(u,v)$ の面積は
$$
\iint_D dS=\iint_D \left|\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\right|\,dudv
$$
に等しい.

#### ベクトル面素 ####

- 曲面の単位法線ベクトル場$\boldsymbol{n}$を「面素（$dS$）倍」したものを, <b>ベクトル面素</b>とよび, $d\boldsymbol{S}$と書く.
- 単位法線ベクトル場$\boldsymbol{n}$は
$$\boldsymbol{n}=\frac{1}{\left|\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\right|}\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}$$
であるから,
$$d\boldsymbol{S}=\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\,dudv$$
である.

### スカラー場の面積分 ###

領域$D$上の関数$\boldsymbol{r}(u,v)$で定義された曲面を$S$とする.

空間のスカラー場$\varphi(x,y,z)$に対し, 2重積分
$$
\iint_D\varphi(\boldsymbol{r}(u,v))\left|\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\right|\,dudv
$$
を, 「$\varphi$の$S$上での面積分」といい, $\displaystyle \int_S\varphi\,dS$ と書く.

<b>注意.</b> スカラー場の面積分は, 曲面のパラメータのとり方に依らない.

この事実を, 半径$1$の上半球面を例として, 確かめる.

#### 例）上半球面上の面積分 ####

i) 原点が中心で半径$1$の球面は, 方程式 $x^2+y^2+z^2=1$ を満たす点$(x,y,z)$の全体である.
よって, 上半球面（$z>0$を満たす部分）は, 関数 $f(x,y)=\sqrt{1-x^2-y^2}$ のグラフ曲面である.

```
HUSfunct[x_,y_]:=Sqrt[1-x^2-y^2];
Plot3D[HUSfunct[x,y],{x,-1,1},{y,-1,1}]
```

法線ベクトル場は

```
Cross[D[{x,y,HUSfunct[x,y]},x],D[{x,y,HUSfunct[x,y]},y]]
```

である. このベクトルの大きさは

```
Simplify[Sqrt[%[[1]]^2+%[[2]]^2+%[[3]]^2]]
```

であるから, 単位法線ベクトル場は

```
{x,y,HUSfunct[x,y]}
```

である.
つまり, 単位法線ベクトル場は, 球面上の点$(x,y,z)$に対し, ベクトル$x\,\boldsymbol{i}+y\,\boldsymbol{j}+z\,\boldsymbol{k}$を対応させるベクトル場である（半径$1$の球面なので, これは明らか）.

ii) 一方, $(\cos v\,\cos u, \cos v\,\sin u, \sin v), 0\leqq u\leqq 1\pi, 0\leqq v\leqq 2\pi$ とパラメータ表示することができる（これは平面の点の極表示の拡張である）.

```
HUSpara[u_,v_]:={Cos[v]*Cos[u], Cos[v]*Sin[u],Sin[v]};
ParametricPlot3D[HUSpara[u,v],{u,0,2*Pi},{v,0,Pi/2}]
```

このパラメータ表示における法線ベクトルは

```
Cross[D[HUSpara[u,v],u],D[HUSpara[u,v],v]]
```

つまり, 
$$
\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}=\cos u\,\cos^2 v\,\boldsymbol{i}+\sin u\,\cos^2 v\,\boldsymbol{j}+\cos v\,\sin v\,\boldsymbol{k}
$$
である. このベクトルの大きさは

```
Sqrt[%[[1]]^2+%[[2]]^2+%[[3]]^2]
```

となり, これは $\cos v$ に等しい. 実際に, 

```
Simplify[%-Cos[v]]
```

<b>スカラー場 $\varphi(x,y,z)=z^2$ の面積分</b>

i) グラフ曲面 $z=\sqrt{1-x^2-y^2}$

- $\varphi\left(x,y,\sqrt{1-x^2-y^2}\right)=1-x^2-y^2$
- $dS=\dfrac{1}{\sqrt{1-x^2-y^2}}\,dxdy$
- 積分領域は $-1\leqq x\leqq 1$, $-\sqrt{1-x^2}\leqq y\leqq\sqrt{1-x^2}$

よって, 面積分の式は
$$
\int^1_{-1}\int^{\sqrt{1-x^2}}_{-\sqrt{1-x^2}}\sqrt{1-x^2-y^2}\,dy\,dx
$$
であり, その値は

```
Integrate[Integrate[HUSfunct[x,y]^2/HUSfunct[x,y],{y,-Sqrt[1-x^2],Sqrt[1-x^2]}],{x,-1,1}]
```

である.

ii) パラメータ表示 $\boldsymbol{r}(u,v)=(\cos v\,\cos u, \cos v\,\sin u, \sin v)$

- $\varphi\left(\boldsymbol{r}(u,v)\right)=\sin^2 v$
- $dS=\cos v\,dudv$
- 積分領域は $0\leqq u\leqq 2\pi$, $0\leqq v\leqq \pi/2$

よって, 面積分の値は
$$
\int^{2\pi}_0\int^{\frac{\pi}{2}}_0 \sin^2v\,\cos v\,dv\,du
$$
であり, その値は

```
Integrate[Integrate[Sin[v]^2*Cos[v],{v,0,Pi/2}],{u,0,2*Pi}]
```

である.

#### スカラー場の面積分の意味 ####

スカラー場が物体の密度分布を表すとすると, 曲面$S$上の面積分は, $S$の表面の質量を表す.

### ベクトル場の面積分 ###

領域$D$上の関数$\boldsymbol{r}(u,v)$で定義された曲面を$S$とする.

空間のベクトル場$\boldsymbol{A}(x,y,z)=A_x(x,y,z)\,\boldsymbol{i}+A_y(x,y,z)\,\boldsymbol{j}+A_z(x,y,z)\,\boldsymbol{k}$に対し, 2重積分
$$
\iint_D \boldsymbol{A}(\boldsymbol{r}(u,v))\cdot\left(\frac{\partial \boldsymbol{r}}{\partial u}\times \frac{\partial \boldsymbol{r}}{\partial v}\right)\,dudv
$$
を, 「$\boldsymbol{A}$の$S$上での面積分」といい, $\displaystyle \int_S\boldsymbol{A}\cdot d\boldsymbol{S}$ と書く.

<b>注意.</b> ベクトル場の面積分の絶対値は, 曲面のパラメータのとり方に依らない（つまり, 面積分の値は, 符号の違いを除いて一意に定まる）.

<b>ベクトル場 $\boldsymbol{A}(x,y,z)=x^2y^2\,\boldsymbol{k}$ の面積分</b>

i) グラフ曲面 $z=\sqrt{1-x^2-y^2}$

- $\boldsymbol{A}\left(x,y,\sqrt{1-x^2-y^2}\right)=x^2y^2\,\boldsymbol{k}$
- $d\boldsymbol{S}=\left(\dfrac{x}{\sqrt{1-x^2-y^2}}\boldsymbol{i}+\dfrac{y}{\sqrt{1-x^2-y^2}}\boldsymbol{j}+\boldsymbol{k}\right)\,dxdy$
- $\boldsymbol{A}\cdot d\boldsymbol{S}=x^2y^2\,dxdy$
- 積分領域は $-1\leqq x\leqq 1$, $-\sqrt{1-x^2}\leqq y\leqq\sqrt{1-x^2}$

よって, 面積分の式は
$$
\int^1_{-1}\int^{\sqrt{1-x^2}}_{-\sqrt{1-x^2}}x^2y^2\,dy\,dx
$$
であり, その値は

```
Integrate[Integrate[x^2*y^2,{y,-Sqrt[1-x^2],Sqrt[1-x^2]}],{x,-1,1}]
```

となる.

ii) パラメータ表示 $\boldsymbol{r}(u,v)=(\cos v\,\cos u, \cos v\,\sin u, \sin v)$

- $\boldsymbol{A}\left(\boldsymbol{r}(u,v)\right)=\cos^4 v\,\cos^2 u\,\sin^2 u\,\boldsymbol{k}$
- $d\boldsymbol{S}=\left(\cos u\,\cos^2 v\,\boldsymbol{i}+\sin u\,\cos^2 v\,\boldsymbol{j}+\cos v\,\sin v\,\boldsymbol{k}\right)\,dudv$
- $\boldsymbol{A}\cdot d\boldsymbol{S}=\cos^5 v\,\sin v\,\cos^2 u\,\sin^2 u\,dudv$
- 積分領域は $0\leqq u\leqq 2\pi$, $0\leqq v\leqq \pi/2$

よって, 面積分のｓ
$$
\int^{2\pi}_{0}\int^{\frac{\pi}{2}}_{0}\cos^5 v \sin v \cos^2 u \sin^2 u\,dv\,du
$$
値は

```
Integrate[Integrate[Cos[v]^5*Sin[v]*Cos[u]^2*Sin[u]^2,{v,0,Pi/2}],{u,0,2*Pi}]
```

となる.

#### ベクトル場の面積分の意味 ####

ベクトル場を非圧縮流体が流れる速度を表すとすると, 曲面$S$上の面積分は, $S$の表面を通過する単位時間あたりの物体の体積を表す.

#### p.100 問題 3 ####

(1) $\displaystyle \int_S (x+y+z)\,dS\quad S : 2x+2y+z=4,\ x\leqq 0,\ y\leqq 0,\ z\leqq 0$

$S$の定義式を$z=4-2x-2y$ と変形すると, これは関数$f(x,y)=4-2x-2y$のグラフ曲面である.
この曲面のグラフは

```
funct31[x_,y_]:=4-2*x-2*y;
Plot3D[funct31[x,y],{x,-2,2},{y,-2,2}]
```

この曲面の法線ベクトルは

```
Cross[{1,0,D[funct31[x,y],x]},{0,1,D[funct31[x,y],y]}]
```

となり, これは定ベクトルである.
このことから, この曲面が平面であることがわかる.
また, 法線ベクトルの成分は, 平面の方程式における$x, y, z$の係数である.

さて, 面積分の積分領域である平面$S$は$x, y, z$がともに正の点の領域である.
各座標軸と平面の交点は, $(2,0,0), (0,2,0), (0,0,4)$なので,
$S$はこの3点を頂点とする三角形領域である.

```
Show[Graphics3D[Polygon[{{2,0,0},{0,2,0},{0,0,4}}]]]
```

これは「$xy$平面における3点$(0,0, (2,0), (0,2)$を頂点とする三角形領域を定義域とする関数$z=4-2x-2y$のグラフ」である.

```
Show[
    Graphics3D[Polygon[{{2,0,0},{0,2,0},{0,0,4}}]],
    Graphics3D[Polygon[{{2,0,0},{0,2,0},{0,0,0}}]]
]
```

定義域の三角形領域は, 不等式
$$
0\leqq x\leqq 2,\quad
0\leqq y\leqq 2-x
$$
または, 
$$
0\leqq x\leqq 2-y,\quad
0\leqq y\leqq 2
$$
と表すことができる.


面素は
$$
dS=\sqrt{2^2+2^2+1^2}dxdy=3\,dxdy
$$

被積分関数は
$$
x+y+z=x+y+(4-2x-2y)=4-x-y
$$

以上のことから,
$$
\int_S (x+y+z)\,dS=\int^2_0\int^{2-x}_0 3(4-x-y)\,dy\,dx
$$
または,
$$
\int_S (x+y+z)\,dS=\int^2_0\int^{2-y}_0 3(4-x-y)\,dx\,dy
$$
であり, 値は

```
Integrate[Integrate[3*(4-x-y),{y,0,2-x}],{x,0,2}]
```

である.
