#### 補足 2.1 ####

ベクトルの外積は, 結合法則
$$
(\boldsymbol{A}\times\boldsymbol{B})\times\boldsymbol{C}=\boldsymbol{A}\times(\boldsymbol{B}\times\boldsymbol{C})
$$
を満たさない.

（例） 3つのベクトル $\boldsymbol{A}, \boldsymbol{B}, \boldsymbol{C}$ を次のようにおく.

```
vecA:= {1,2,3};
vecB:= {1,-1,3};
vecC:={2,1,2};
```

このとき, $(\boldsymbol{A}\times\boldsymbol{B})\times\boldsymbol{C}$ は

```
Cross[Cross[vecA, vecB],vecC]
```

一方, $\boldsymbol{A}\times(\boldsymbol{B}\times\boldsymbol{C})$ は

```
Cross[vecA, Cross[vecB,vecC]]
```

このように, 一般に, $(\boldsymbol{A}\times\boldsymbol{B})\times\boldsymbol{C}\ne\boldsymbol{A}\times(\boldsymbol{B}\times\boldsymbol{C})$ であることがわかる.

### 第5回 5月16日（木） スカラー場と勾配 ###

<span style="background-color: forestgreen">&nbsp;第2部 第3章 &#167;3.1&nbsp;</span>
 
 <ul>
    <li>微分演算子（<b>ナブラ演算子</b>） $\nabla=\dfrac{\partial}{\partial x}\,\boldsymbol{i}+\dfrac{\partial}{\partial y}\,\boldsymbol{j}+\dfrac{\partial}{\partial z}\,\boldsymbol{k}$</li>
    <li><b>方向微分係数</b>
    <ul>
        <li>点$\mathrm{P}(x,y,z)$ を始点とする単位ベクトル $\boldsymbol{v}$ とスカラー場 $\varphi(x,y,z)$ から定まる値; $\dfrac{d\varphi}{d\boldsymbol{v}}=\nabla \varphi(\mathrm{P})\cdot \boldsymbol{v}$</li>
        <li>$\dfrac{d\varphi}{d\boldsymbol{v}}\approx\dfrac{\varphi({\rm P}+h\boldsymbol{v})-\varphi({\rm P})}{t}$
    </ul></li>
    <li>スカラー場 $\varphi$ の<b>等位面</b>
        <ul>
            <li>$\varphi(x,y,z)=c$ を満たす点$(x,y,z)$の全体のなす曲面のこと.</li>
            <li>勾配 $\nabla\varphi$ は, 等位面 $\varphi(x,y,z)=c$ に垂直なベクトルである. つまり, $\nabla\varphi$ は等位面の単位法線ベクトル場 $\boldsymbol{n}$ と平行である. 特に, $\nabla \varphi=\dfrac{d\varphi}{d\boldsymbol{n}}\boldsymbol{n}$ であり, $|\nabla\varphi|=\dfrac{d\varphi}{d\boldsymbol{n}}$.</li>
            <li>事実：等位面が密なところでは $|\nabla\varphi|$ は大きく, 疎なところでは $|\nabla\varphi|$ は小さい.<br>
            $\because$ 2つの等位面 $\varphi(x,y,z)=c_1$ と $\varphi(x,y,z)=c_2$ から2点 ${\rm P}, {\rm Q}={\rm P}+t\boldsymbol{n}$ をそれぞれ選ぶと,<br>
            $$|\nabla\varphi({\rm P})|=\dfrac{d\varphi}{d\boldsymbol{n}}({\rm P})\approx\dfrac{\varphi({\rm P}+t\boldsymbol{n})-\varphi({\rm P})}{t}=\dfrac{c_2-c_1}{t}$$
            よって, 等位面が密（疎） $\Longleftrightarrow$ $t$が小（大） $\Longleftrightarrow$ $|\nabla\varphi|$ が大（小）</li>
        </ul>
    </li>
</ul>

### 第7回 5月30日（木）空間図形(1) 曲線 ###

<span style="background-color: forestgreen">&nbsp;第2部 第3章 &#167;3.3&nbsp;</span>

- <b>曲線の媒介変数表示（パラメータ表示）</b><br>
ベクトル関数 $\boldsymbol{r}(t)=x(t)\,\boldsymbol{i}+y(t)\,\boldsymbol{j}+z(t)\,\boldsymbol{k}$ は, $t$ に対して空間内の点 $(x(t), y(t), z(t))$ を対応させる関数とみなすことができる. このような点の全体は空間内の曲線をなす（以後, ベクトル関数を空間内の曲線を同一視する）.
- <b>曲線の接ベクトル</b><br>
ベクトル関数 $\boldsymbol{r}(t)$ が定める曲線を $C$ とする.
このとき, $\boldsymbol{r}(t)$ の微分 $\boldsymbol{r}'(t)=x'(t)\,\boldsymbol{i}+y'(t)\,\boldsymbol{j}+z'(t)\,\boldsymbol{k}$ を「点 $\boldsymbol{r}(t)$ における曲線 $C$ の接ベクトル」という.
- <b>曲線の接ベクトル場</b><br>
曲線 $C$ の各点 $\boldsymbol{r}(t_0)$ に, ベクトル$\boldsymbol{r}'(t_0)$ を対応させる関数を「曲線 $C$ 上の接ベクトル場」という.
- <b>曲線の弧長</b><br>
曲線 $C$ の $a\leqq t\leqq b$ における長さは
$$
\int^b_a|\boldsymbol{r}'(t)|\,dt=\int^b_z\sqrt{x'(t)^2+y'(t)^2+z'(t)^2}\,dt
$$
で与えられる.

#### 弧長を求める方法 ####

曲線の弧長を次の方法で定義した;

- 曲線 $C$ を $n$ 分割する.
- 隣り合う2点を端点とする線分の長さを求め, $n$ 個の総和 $\ell_n$ を求める.
- $n\rightarrow \infty$ としたときの $\ell_n$ の極限値を弧長と定める.

これは, 曲線を折れ線の「極限」と考えることに他ならない.<br>
※ 厳密には, 「どのような分割の仕方に対しても,  $\ell_n$ の極限がある一定値になるとき, その極限」を弧長と定める（ことができる）.

この方法にしたがって, $xy$-平面内の半径1の円周の長さを求めてみよう.
中心を原点とすると, この曲線は $\boldsymbol{r}(t)=\cos t\,\boldsymbol{i}+\sin t\,\boldsymbol{j},\ 0\leqq t\leqq 2\pi$ と書ける.
定義域を $n$ 等分すると, 曲線上の分点
$$
(1,0,0),\ldots,\left(\cos \frac{2i\pi}{n},\sin \frac{2i\pi}{n},0\right),\ldots,\left(\cos \frac{2(n-1)\pi}{n},\sin \frac{2(n-1)\pi}{n},0\right),\ (1,0,0)
$$
を得る.
隣り合う2点を端点とする線分の長さは
\begin{align}
&\sqrt{\left\{\cos \frac{2i\pi}{n}-\cos \frac{2(i-1)\pi}{n}\right\}^2+\left\{\sin \frac{2i\pi}{n}-\sin \frac{2(i-1)\pi}{n}\right\}^2}\\
=&\sqrt{2-2\cos \frac{2i\pi}{n}\,\cos \frac{2(i-1)\pi}{n}-2\sin \frac{2i\pi}{n}\,\sin \frac{2(i-1)\pi}{n}}\\
=&\sqrt{2-2\cos\left(\frac{2i\pi}{n}-\frac{2(i-1)\pi}{n}\right)}\quad\because\mbox{（加法定理）}\\
=&\sqrt{2-2\cos\frac{2\pi}{n}}\\
=&\sqrt{2-2\left(1-2\sin^2\frac{\pi}{n}\right)}\quad\because\mbox{（半角の公式）}\\
=&\sqrt{4\sin^2\frac{\pi}{n}}\\
=&2\sin\frac{\pi}{n}
\end{align}
となる（この値が $i$ に依らないことに注意）.
よって, 折れ線（正$n$角形）の長さの総和は
$$
2n\sin\frac{\pi}{n}
$$
である．
この極限値は

```
Limit[2*n*Sin[Pi/n], n->Infinity]
```

となる.<br>

#### 小テスト⑥についての補足 ####

- ベクトル場 $\boldsymbol{A}$ はスカラー場 $\varphi_A(x,y,z)=x^2\log y - x^3z^2$ の勾配ベクトル場である. よって, $\boldsymbol{A}$ の発散は $\varphi_A$ のラプラシアンであり, $\boldsymbol{A}$ の回転は零ベクトルの定ベクトル場である.
- ベクトル場 $\boldsymbol{B}$ はベクトル場 $\boldsymbol{C}(x,y,z)=xyz\boldsymbol{i}-y^2z^3\boldsymbol{j}+x^2y\boldsymbol{k}$ の回転ベクトル場である. よって, $\boldsymbol{B}$ の発散は定値スカラー場 $0$ である.
- 勾配ベクトル場の発散はラプラシアン $\Delta$ であり, $\Delta \varphi=0$ であることから, $\varphi$ は調和関数であることがわかる.
