# ベクトル解析 <span style="color: #808080; ">Vector Analysis</span> #

## 曲面 ##

2つの独立変数 $u, v$ をもつベクトル関数は, 空間内の曲面を表す.

```
JupyterImageResolution = 72;
```

例1） $\boldsymbol{r}(u,v)=(u+v)\,\boldsymbol{i}+(y-v)\,\boldsymbol{j}+\sin(uv)\,\boldsymbol{k}$

```
surfpara1[u_,v_]:={u+v,u-v,Sin[u*v]};
```

```
ParametricPlot3D[surfpara1[u,v],{u,-2,2},{v,-2,2}]
```

例2） $\boldsymbol{r}(u,v)=u \cos v\,\boldsymbol{i}+u \sin \,\boldsymbol{j}+(1-u^2)\cos u\,\boldsymbol{k}$

```
surfpara2[u_,v_]:={u*Cos[v],u*Sin[v],(-u^2+1)*Cos[u]};
```

```
ParametricPlot3D[surfpara2[u,v],{u,0,2},{v,0,2*Pi}]
```

例3）2変数関数 $f(x,y)=\sqrt{x^2+y^2}$ が定めるグラフ曲面

```
surfpara3[x_,y_]:={x,y,Sqrt[x^2+y^2]};
```

```
ParametricPlot3D[surfpara3[x,y],{x,-1,1},{y,-1,1}]
```

### 曲面の座標曲線 ###

空間内の曲面 $\boldsymbol{r}(u,v)=(x(u,v), y(u,v), z(u,v))$ に対し,
$v=v_0$ に固定すると, $\boldsymbol{r}(u,v_0)$ は $u$ をパラメータとする曲線となる.
これを「曲面 $\boldsymbol{r}(u,v)$ の $u$曲線」という.

同様に, 曲面 $\boldsymbol{r}(u,v)$ の $v$曲線を考えることができる.

例1） $\boldsymbol{r}(u,v)=(u+v)\,\boldsymbol{i}+(y-v)\,\boldsymbol{j}+\sin(uv)\,\boldsymbol{k}$

```
Show[
    ParametricPlot3D[surfpara1[u,v],{u,-2,2},{v,-2,2},Mesh->False],
    ParametricPlot3D[surfpara1[u,1],{u,-2,2}]
]
```

```
Show[
    ParametricPlot3D[surfpara1[u,v],{u,-2,2},{v,-2,2},Mesh->False],
    ParametricPlot3D[surfpara1[1,v],{v,-2,2}]
]
```

例2） $\boldsymbol{r}(u,v)=u \cos v\,\boldsymbol{i}+u \sin \,\boldsymbol{j}+(1-u^2)\cos u\,\boldsymbol{k}$

```
Show[
    ParametricPlot3D[surfpara2[u,v],{u,0,2},{v,0,2*Pi},Mesh->False],
    ParametricPlot3D[surfpara2[u,3*Pi/2],{u,0,1}]
]
```

```
Show[
    ParametricPlot3D[surfpara2[u,v],{u,0,2},{v,0,2*Pi},Mesh->False],
    ParametricPlot3D[surfpara2[2/3,v],{v,0,2*Pi}]
]
```

### 偏微分$\boldsymbol{r}_u(u,v)$と$\boldsymbol{r}_v(u,v)$の外積 ###

- 曲面 $\boldsymbol{r}(u,v)$ の偏微分は, 座標曲線に接するベクトルと考えることができる.
- 通常，曲面を扱う際は,
$$
\boldsymbol{r}_u(u,v)\times\boldsymbol{r}_v(u,v)\ne \boldsymbol{0}
$$
を仮定する.
- 「$\boldsymbol{r}_u(u,v)\times\boldsymbol{r}_v(u,v)=\boldsymbol{0}$」$\Longleftrightarrow$「$\boldsymbol{r}_u(u,v)=\boldsymbol{0}$」または「$\boldsymbol{r}_v(u,v)=\boldsymbol{0}$」または「$\boldsymbol{r}_u(u,v)$と$\boldsymbol{r}_v(u,v)$が平行」

### 曲面の接平面 ###

- 点 $\boldsymbol{r}(a,b)$ を通り, 2つのベクトル$\boldsymbol{r}_u(a,b), \boldsymbol{r}_v(a,b)$ を含む平面を「点$\boldsymbol{r}(a,b)$における曲面$\boldsymbol{r}(u,v)$の<b>接平面</b>」という.
- <b>接平面の方程式</b>：点$\boldsymbol{r}(a,b)$における曲面$\boldsymbol{r}(u,v)$の接平面上の点 $(x,y,z)$ は
$$
\boldsymbol{r}_u(a,b)\times\boldsymbol{r}_v(a,b)\cdot (x\,\boldsymbol{i}+y\,\boldsymbol{j}+z\,\boldsymbol{k}-\boldsymbol{r}(a,b))=0
$$
を満たす.
- <b>接平面のパラメータ表示</b>：
$$
\boldsymbol{r}(a,b)+t\,\boldsymbol{r}_u(a,b)+s\,\boldsymbol{r}_v(a,b)
$$

### 曲面の法線ベクトル ###

- ベクトル $\boldsymbol{r}_u(a,b)\times\boldsymbol{r}_v(a,b)$ は, 接平面に垂直なベクトルである. このようなベクトルを曲面$\boldsymbol{r}$の点$\boldsymbol{r}(a,b)$における<b>法線ベクトル</b>という.
- 特に, 
$$
\frac{1}{|\boldsymbol{r}_u(a,b)\times\boldsymbol{r}_v(a,b)|}
\boldsymbol{r}_u(a,b)\times\boldsymbol{r}_v(a,b)
$$
を<b>単位法線ベクトル</b>, または<b>法単位ベクトル</b>という.
- 曲面上の各点に対し, その点における（単位）法線ベクトルを対応させるベクトル場を（<b>単位</b>）<b>法線ベクトル場</b>という.

例1） $\boldsymbol{r}(u,v)=(u+v)\,\boldsymbol{i}+(y-v)\,\boldsymbol{j}+\sin(uv)\,\boldsymbol{k}$

法線ベクトル場は

```
Cross[D[surfpara1[u,v],u],D[surfpara1[u,v],v]]
```

```
normalvecf1[u_,v_]:=Cross[D[surfpara1[x,y],x],D[surfpara1[x,y],y]]/.x->u/.y->v
```

```
normalvec1=normalvecf1[1,1];
tangentvec1u=D[surfpara1[u,v],u]/.u->1/.v->1;
tangentvec1v=D[surfpara1[u,v],v]/.u->1/.v->1;
```

```
D[surfpara1[u,v],u]/.u->1/.v->1
```

```
D[surfpara1[u,v],v]/.u->1/.v->1
```

```
Show[
    Graphics3D[{PointSize[0.03],Point[surfpara1[1,1]]}],
    ParametricPlot3D[surfpara1[u,v],{u,-2,2},{v,-2,2},Mesh->False,PlotStyle->Opacity[0.6]],
    ParametricPlot3D[surfpara1[u,1],{u,-2,2}],
    ParametricPlot3D[surfpara1[1,v],{v,-2,2}],
    Graphics3D[{Thick,Arrow[{surfpara1[1,1],surfpara1[1,1]+normalvec1}]}],
    ParametricPlot3D[surfpara1[1,1]+t*tangentvec1u+s*tangentvec1v,{t,-1,1},{s,-1,1}]
]
```

```
Show[
    ParametricPlot3D[surfpara1[u,v],{u,-2,2},{v,-2,2},Mesh->False,PlotStyle->Opacity[0.6]],
    Table[
    Graphics3D[{Arrowheads[0.02],
    Table[Arrow[{surfpara1[u,v],surfpara1[u,v]-normalvecf1[u,v]/Norm[normalvecf1[u,v]]}],
          {v,-2,2,2/5}]}
    ],{u,-2,2,2/5}
    ]
]
```

例2） $\boldsymbol{r}(u,v)=u \cos v\,\boldsymbol{i}+u \sin \,\boldsymbol{j}+(1-u^2)\cos u\,\boldsymbol{k}$

法線ベクトル場は

```
Simplify[Cross[D[surfpara2[u,v],u],D[surfpara2[u,v],v]]]
```

```
normalvecf2[u_,v_]:=Simplify[Cross[D[surfpara2[x,y],x],D[surfpara2[x,y],y]]]/.x->u/.y->v
```

```
normalvec2=normalvecf2[2/3,3*Pi/2];
tangentvec2u=D[surfpara2[u,v],u]/.u->2/3/.v->3*Pi/2;
tangentvec2v=D[surfpara2[u,v],v]/.u->2/3/.v->3*Pi/2;
```

```
Show[
    Graphics3D[{PointSize[0.03],Point[surfpara2[2/3,3*Pi/2]]}],
    ParametricPlot3D[surfpara2[u,v],{u,0,2},{v,0,2*Pi},Mesh->False,PlotStyle->Opacity[0.6]],
    ParametricPlot3D[surfpara2[u,3*Pi/2],{u,0,2}],
    ParametricPlot3D[surfpara2[2/3,v],{v,0,2*Pi}],
    Graphics3D[{Thick,Arrow[{surfpara2[2/3,3*Pi/2],surfpara2[2/3,3*Pi/2]+normalvec2}]}],
    ParametricPlot3D[surfpara2[2/3,3*Pi/2]+t*tangentvec2u+s*tangentvec2v,{t,-1,1},{s,-1,1}]
]
```

この曲面は, $u=0$ のとき, 法線ベクトルが零ベクトルになる.

```
normalvecf2[0,v]
```

しかし, 単位法線ベクトルは, $u\rightarrow 0$ のとき, ベクトル $\boldsymbol{k}$ に収束する.

```
Limit[Simplify[normalvecf2[u,v]/Sqrt[normalvecf2[u,v][[1]]^2+normalvecf2[u,v][[2]]^2+normalvecf2[u,v][[3]]^2],u>0],u->0]
```

```
Show[ParametricPlot3D[surfpara2[u,v],{u,0,2},{v,0,2*Pi},Mesh->False,PlotStyle->Opacity[0.6]],
    Table[Graphics3D[{Arrowheads[0.02],
    Table[Arrow[{surfpara2[u,v],surfpara2[u,v]+normalvecf2[u,v]/Norm[normalvecf2[u,v]]}],
          {v,0,2*Pi,2*Pi/5}]}],{u,0+0.001,2,1/5}]]
```

例3）2変数関数 $f(x,y)=\sqrt{x^2+y^2}$ が定めるグラフ曲面

法線ベクトル場は

```
Cross[D[surfpara3[x,y],x],D[surfpara3[x,y],y]]
```

```
normalvecf3[x_,y_]:=Simplify[Cross[D[surfpara3[u,v],u],D[surfpara3[u,v],v]]]/.u->x/.v->y
```

この曲面の法線ベクトルは, 大きさが一定である.

```
Sqrt[Simplify[normalvecf3[x,y][[1]]^2+normalvecf3[x,y][[2]]^2+normalvecf3[x,y][[3]]^2]]
```

```
normalvec3=normalvecf3[1/2,1/2];
tangentvec3x=D[surfpara3[x,y],x]/.x->1/2/.y->1/2;
tangentvec3y=D[surfpara3[x,y],y]/.x->1/2/.y->1/2;
```

```
Show[
    Graphics3D[{PointSize[0.03],Point[surfpara3[1/2,1/2]]}],
    ParametricPlot3D[surfpara3[x,y],{x,-1,1},{y,-1,1},Mesh->False,PlotStyle->Opacity[1.0]],
    ParametricPlot3D[surfpara3[x,1/2],{x,-1,1}],
    ParametricPlot3D[surfpara3[1/2,y],{y,-1,1}],
    Graphics3D[{Thick,Arrow[{surfpara3[1/2,1/2],surfpara3[1/2,1/2]+normalvec3}]}],
    ParametricPlot3D[surfpara3[1/2,1/2]+t*tangentvec3x+s*tangentvec3y,{t,-1,1},{s,-1,1}]
]
```

```
Show[ParametricPlot3D[surfpara3[x,y],{x,-1,1},{y,-1,1},Mesh->False,PlotStyle->Opacity[1]],
    Table[Graphics3D[{Arrowheads[0.02],
    Table[Arrow[{surfpara3[x,y],surfpara3[x,y]+normalvecf3[x,y]/Sqrt[2]}],
          {x,0.01,1,1/10}]}],{y,-1,1,1/5}]]
```
