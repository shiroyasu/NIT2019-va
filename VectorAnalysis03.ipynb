### スカラー場とベクトル場 ###

- 空間内の点 $(x,y,z)$ に対し, スカラー $\varphi(x,y,z)$ が対応しているとき, この対応のことを<b>スカラー場</b>とよぶ.
- スカラー場とは, 3変数関数のことである.

- 空間内の点 $(x,y,z)$ に対し, ベクトル $A_x(x,y,z)\,\boldsymbol{i}+A_y(x,y,z)\,\boldsymbol{j}+A_z(x,y,z)\,\boldsymbol{k}$ が対応しているとき, この対応のことを<b>ベクトル場</b>とよぶ.
- ベクトル場とは, 3変数のベクトル関数, または3変数関数の3つの組のことである.

### スカラー場の勾配 ###

スカラー場 $\varphi(x,y,z)$ に対し, 次式で定まるベクトル場を $\varphi$ の勾配 (gradient) という;
$$
\mathrm{grad}(\varphi)=\dfrac{\partial\varphi}{\partial x}\,\boldsymbol{i}+\dfrac{\partial\varphi}{\partial y}\,\boldsymbol{j}+\dfrac{\partial\varphi}{\partial z}\,\boldsymbol{k}
$$

#### p.84 例題2 ####

$\varphi(x,y,z)=x^2\log y-x^3z^2$ の勾配を求める.

```
Grad[x^2*Log[y]-x^3*z^2,{x,y,z}]
```

```
VectorPlot3D[{-3 x^2 z^2 + 2 x Log[y], x^2/y, -2 x^3 z}, {x, 1,5}, {y, 1, 5}, {z, -2, 2}, VectorPoints -> Fine]
```

### 方向微分係数と勾配 ###

- 点$\mathrm{P}(x,y,z)$ を始点とする単位ベクトル $\boldsymbol{v}$ とスカラー場 $\varphi(x,y,z)$ から定まる値; $\dfrac{d\varphi}{d\boldsymbol{v}}=\nabla \varphi(\mathrm{P})\cdot \boldsymbol{v}$
- $\dfrac{d\varphi}{d\boldsymbol{v}}\approx\dfrac{\varphi({\rm P}+h\boldsymbol{v})-\varphi({\rm P})}{t}$


### スカラー場の等位面 ###

- $\varphi(x,y,z)=c$ を満たす点$(x,y,z)$の全体のなす曲面のこと.
- 勾配 $\nabla\varphi$ は, 等位面 $\varphi(x,y,z)=c$ に垂直なベクトルである. つまり, $\nabla\varphi$ は等位面の単位法線ベクトル場 $\boldsymbol{n}$ と平行である. 特に, $\nabla \varphi=\dfrac{d\varphi}{d\boldsymbol{n}}\boldsymbol{n}$ であり, $|\nabla\varphi|=\dfrac{d\varphi}{d\boldsymbol{n}}$.
- 事実：等位面が密なところでは $|\nabla\varphi|$ は大きく, 疎なところでは $|\nabla\varphi|$ は小さい.

$\because$ 2つの等位面 $\varphi(x,y,z)=c_1$ と $\varphi(x,y,z)=c_2$ から2点 ${\rm P}, {\rm Q}={\rm P}+t\boldsymbol{n}$ をそれぞれ選ぶと,
$$|\nabla\varphi({\rm P})|=\dfrac{d\varphi}{d\boldsymbol{n}}({\rm P})\approx\dfrac{\varphi({\rm P}+t\boldsymbol{n})-\varphi({\rm P})}{t}=\dfrac{c_2-c_1}{t}$$
よって, 等位面が密（疎） $\Longleftrightarrow$ $t$が小（大） $\Longleftrightarrow$ $|\nabla\varphi|$ が大（小）

#### ベクトル場の発散（p.88 問題1.(1)） ###

ベクトル場 $\boldsymbol{A}(x,y,z)=x^3z^2\,\boldsymbol{i}+(xy^2+z^2)\,\boldsymbol{j}-x^2z^3\,\boldsymbol{k}$ の発散とその等位面をみてみる.

```
VectorPlot3D[{x^3*z^2,x*y^2+z^2,-x^2*z^3}, {x, -5,5}, {y, -5, 5}, {z, -5, 5}, VectorPoints -> Fine]
```

```
Div[{x^3*z^2,x*y^2+z^2,-x^2*z^3},{x,y,z}]
```

```
ContourPlot3D[2*x*y,{x,-5,5},{y,-5,5},{z,-5,5},Contours->10]
```

```
Show[
    ContourPlot3D[2*x*y,{x,-5,5},{y,-5,5},{z,-5,5},Contours->10],
    VectorPlot3D[{x^3*z^2,x*y^2+z^2,-x^2*z^3}, {x, -5,5}, {y, -5, 5}, {z, -5, 5}, VectorPoints -> Fine]
]
```

#### ベクトル場の回転（p.88 問題3.(1)） ###

ベクトル場 $\boldsymbol{A}(x,y,z)=xyz\,\boldsymbol{i}-y^2z^3\,\boldsymbol{j}+2x^2y\,\boldsymbol{k}$ とその回転との関係をみてみる.

```
VectorPlot3D[{x*y*z,-y^2*z^3,2*x^2*y}, {x, -5,5}, {y, -5, 5}, {z, -5, 5}, VectorPoints -> Fine]
```

```
Curl[{x*y*z,-y^2*z^3,2*x^2*y},{x,y,z}]
```

```
VectorPlot3D[%, {x, -5,5}, {y, -5, 5}, {z, -5, 5}, VectorPoints -> Fine, VectorStyle -> Red]
```

```
Show[
    VectorPlot3D[{x*y*z,-y^2*z^3,2*x^2*y}, {x, -5,5}, {y, -5, 5}, {z, -5, 5}, VectorPoints -> Fine],
    VectorPlot3D[{2*x^2+3*y^2*z^2,-3*x*y,-x*z}, {x, -5,5}, {y, -5, 5}, {z, -5, 5}, VectorPoints -> Fine, VectorStyle -> Red]
]
```
