# ベクトル解析 <span style="color: #808080; ">Vector Analysis</span> #

## 曲面 ##

2つの独立変数　$u, v$ をもつベクトル関数は, 空間内の曲面を表す.

```
JupyterImageResolution = 72;
```

例1） $\boldsymbol{r}(u,v)=(u+v)\,\boldsymbol{i}+(y-v)\,\boldsymbol{j}+\sin(uv)\,\boldsymbol{k}$

```
ParametricPlot3D[{u+v,u-v,Sin[u*v]},{u,-2,2},{v,-2,2}]
```

例2） $\boldsymbol{r}(u,v)=u \cos v\,\boldsymbol{i}+u \sin \,\boldsymbol{j}+(1-u^2)\cos u\,\boldsymbol{k}$

```
ParametricPlot3D[{u*Cos[v],u*Sin[v],(-u^2+1)*Cos[u]},{u,0,1},{v,0,2*Pi}]
```

例3）2変数関数 $f(x,y)=\sqrt{x^2+y^2}$ が定めるグラフ曲面

```
ParametricPlot3D[{x,y,Sqrt[x^2+y^2]},{x,-1,1},{y,-1,1}]
```

### 曲面の座標曲線と偏微分 ###

空間内の曲面 $\boldsymbol{r}(u,v)=(x(u,v), y(u,v), z(u,v))$ に対し,
$v=v_0$ に固定すると, $\boldsymbol{r}(u,v_0)$ は $u$ をパラメータとする曲線となる.
これを「曲面 $\boldsymbol{r}(u,v)$ の $u$曲線」という.

同様に, 曲面 $\boldsymbol{r}(u,v)$ の $v$曲線を考えることができる.

例1） $\boldsymbol{r}(u,v)=(u+v)\,\boldsymbol{i}+(y-v)\,\boldsymbol{j}+\sin(uv)\,\boldsymbol{k}$

```
Show[
    ParametricPlot3D[{u+v,u-v,Sin[u*v]},{u,-2,2},{v,-2,2},Mesh->False],
    ParametricPlot3D[{u+v,u-v,Sin[u*v]}/.v->1,{u,-2,2}]
]
```

```
Show[
    ParametricPlot3D[{u+v,u-v,Sin[u*v]},{u,-2,2},{v,-2,2},Mesh->False],
    ParametricPlot3D[{u+v,u-v,Sin[u*v]}/.u->1,{v,-2,2}]
]
```

例2） $\boldsymbol{r}(u,v)=u \cos v\,\boldsymbol{i}+u \sin \,\boldsymbol{j}+(1-u^2)\cos u\,\boldsymbol{k}$

```
Show[
    ParametricPlot3D[{u*Cos[v],u*Sin[v],(-u^2+1)*Cos[u]},{u,0,1},{v,0,2*Pi},Mesh->False],
    ParametricPlot3D[{u*Cos[v],u*Sin[v],(-u^2+1)*Cos[u]}/.v->3*Pi/2,{u,0,1}]
]
```

```
Show[
    ParametricPlot3D[{u*Cos[v],u*Sin[v],(-u^2+1)*Cos[u]},{u,0,1},{v,0,2*Pi},Mesh->False],
    ParametricPlot3D[{u*Cos[v],u*Sin[v],(-u^2+1)*Cos[u]}/.u->2/3,{v,0,2*Pi}]
]
```

$u$曲線の微分は, 曲線に接するベクトルであった.
これは $\boldsymbol{r}(u,v)$ を偏微分したベクトルである.

例1） $\boldsymbol{r}(u,v)=(u+v)\,\boldsymbol{i}+(y-v)\,\boldsymbol{j}+\sin(uv)\,\boldsymbol{k}$

```
Show[
    Graphics3D[{PointSize[0.03],Point[{u+v,u-v,Sin[u*v]}/.u->1/.v->1]}],
    ParametricPlot3D[{u+v,u-v,Sin[u*v]},{u,-2,2},{v,-2,2},Mesh->False],
    ParametricPlot3D[{u+v,u-v,Sin[u*v]}/.v->1,{u,-2,2}],
    ParametricPlot3D[{u+v,u-v,Sin[u*v]}/.u->1,{v,-2,2}]
]
```
